```python
from sklearn.preprocessing import OneHotEncoder

enc_soil_type = OneHotEncoder()
enc_soil_type.fit_transform(df_train['Soil Type'])

enc_crop_type = OneHotEncoder()
enc_crop_type.fit_transform(df_train['Crop Type'])

enc_fertilizer_name = OneHotEncoder()
enc_fertilizer_name.fit_transform(df_train['Fertilizer Name'])
```

```python
def preprocessing(df):
    df = df.copy()
    df = df.drop(columns=['id'])
    df.columns = df.columns.str.lower().str.replace(' ', '_')

    df['soil_type'] = enc_soil_type.transform(df['soil_type'])
    df['crop_type'] = enc_crop_type.transform(df['crop_type'])
    if 'fertilizer_name' in df.columns:
        df['fertilizer_name'] = enc_fertilizer_name.transform(df['fertilizer_name'])

    return df

y = preprocessing(df_train)['fertilizer_name'].values
X = preprocessing(df_train).drop(columns=['fertilizer_name']).values
X_test = preprocessing(df_test)
```
